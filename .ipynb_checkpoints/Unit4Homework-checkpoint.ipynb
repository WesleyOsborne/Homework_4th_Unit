```python
# Initial imports
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
import seaborn as sns

%matplotlib inline
```

```python
# Read all Csv Portfolios
Whale_Returns = pd.read_csv("whale_returns.csv", index_col="Date", infer_datetime_format=True, parse_dates=True)
Algorithmic_Returns = pd.read_csv("algo_returns.csv", index_col="Date", infer_datetime_format=True, parse_dates=True)
SP_TSX60 = pd.read_csv("sp_tsx_history.csv", index_col="Date", infer_datetime_format=True, parse_dates=True)
```

```python
# Cleaning Whale Returns (Removing Nulls)
Whale_Returns.isnull().sum()
```

```python
Whale_Returns.dropna(inplace=True)
```

```python
# Cleaning Algorithmic Returns
Algorithmic_Returns.isnull().sum()
```

```python
Algorithmic_Returns.dropna(inplace=True)
Algorithmic_Returns.head()
```

```python
# Cleaning S&P TSX 60
# Fix data to remove dollar signs and comma's
col_SP_TSX60 = SP_TSX60.columns
SP_TSX60[col_SP_TSX60] = SP_TSX60[col_SP_TSX60].replace({r'\$':''}, regex = True)
SP_TSX60[col_SP_TSX60] = SP_TSX60[col_SP_TSX60].replace({r',':''}, regex = True)
SP_TSX60.head()
```

```python
# Make S&P Data float to calculate daily returns
SP_TSX60=SP_TSX60.astype(float)
```

```python
SP_TSX60_daily_returns = SP_TSX60.pct_change()
SP_TSX60_returns = SP_TSX60_daily_returns.rename(columns = {"Close": "S&P TSX 60 Index"})
SP_TSX60_returns.head()
```

```python
# Remove Nulls for S&P Portfolio
SP_TSX60_returns.isnull().sum()
```

```python
SP_TSX60_returns.dropna(inplace=True)
```

```python
# Join Whale Returns, Algorithmic Returns and S&P TSX 60 Returns
Combine_Portfolio_Daily_Returns = pd.concat([SP_TSX60_returns, Whale_Returns, Algorithmic_Returns], axis='columns', join='inner')
```

```python
Combine_Portfolio_Daily_Returns.head()
```

```python
Combine_Portfolio_Daily_Returns.plot(figsize=(20,10), title = "Daily Returns")
```

```python
# Calculate cumulative returns of all portfolios
Cumulative_Returns = (1+Combine_Portfolio_Daily_Returns).cumprod()
Cumulative_Returns
```

```python
Cumulative_Returns.plot(figsize=(20,10), title = "Cumulative Returns")
```

```python
# Box plot to visually show risk
Combine_Portfolio_Daily_Returns.plot.box(title = "Portfolio Risk", figsize=(20,10))
```

```python
# Calculate Daily Standard Deviation of all portfolios
Daily_STD = Combine_Portfolio_Daily_Returns.std()
Daily_STD
```

```python
# Determine What Portfolios are Riskier than the S&P TSX 60 
Riskier_Portfolios = Daily_STD > Daily_STD ['S&P TSX 60 Index']
Riskier_Portfolios
```

```python
# Calculate the annualized standard deviation (252 trading days)
import math
Annual_STD = Daily_STD*math.sqrt(252)
Annual_STD
```

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
Combine_Portfolio_Daily_Returns.rolling(window=21).std()
```

```python
Combine_Portfolio_Daily_Returns.rolling(window=21).std().plot(figsize=(25,10))
```

```python
Combine_21Day_Correlation = Combine_Portfolio_Daily_Returns.corr()
Combine_21Day_Correlation
```

```python
sns.heatmap(Combine_21Day_Correlation, annot=True)
```

```python
# Calculate Variance of S&P TSX 60
SP_TSX60_Variance = Combine_Portfolio_Daily_Returns["S&P TSX 60 Index"].rolling(window=60).var()
SP_TSX60_Variance
```

```python
# Calculate Covariance of Algo 1 portfolio
Algo1_Covariance = Combine_Portfolio_Daily_Returns["Algo 1"].rolling(window=60).cov(Combine_Portfolio_Daily_Returns["S&P TSX 60 Index"])
Algo1_Covariance
```

```python
# Computing Beta
Beta = Algo1_Covariance / SP_TSX60_Variance
Beta
```

```python
#Plot Beta Trend
Beta.plot(figsize=(20,10), title = 'Algo 1 Beta')
```

```python
# Calculate EWM with 21 day halflife
Rolling_EWM = Combine_Portfolio_Daily_Returns.ewm(halflife=21, adjust=True).std().plot(figsize=(20,10), title="Exponential Weighted Average")
```

```python
# Calculate Sharpe Variables
Sharpe_Ratio = Combine_Portfolio_Daily_Returns.mean()/ Combine_Portfolio_Daily_Returns.std()
Sharpe_Ratio
```

```python
# Calculate Annualized Sharpe Variables
Annualized_Sharpe_Ratio = Sharpe_Ratio * math.sqrt(252)
Annualized_Sharpe_Ratio
```

```python
# Visualize the Sharpe Ratios as a bar plot
import matplotlib.pyplot as plt

Annualized_Sharpe_Ratio.plot(kind="bar", title="Annualized Sharpe Ratios")
```

```python
# Based on the Bar Graph we can conclude that both Algorithm 1 and 2 outperformed the S&P TSX 60
# Algorithm 1 outperformed all other portfolios
# Algorithm 2 outperformed all other portfolios except Berkshire Hathaway Inc.
```

```python
# Read the Custon Portfolios
Shop_Returns = pd.read_csv("shop_historical.csv", index_col="Date", infer_datetime_format=True, parse_dates=True)
L_Returns = pd.read_csv("l_historical.csv", index_col="Date", infer_datetime_format=True, parse_dates=True)
Otex_Returns = pd.read_csv("otex_historical.csv", index_col="Date", infer_datetime_format=True, parse_dates=True)
```

```python
# Rename the portfolios, changing close to stock name.
Shop_Returns = Shop_Returns.rename(columns = {"Close": "Shop"})
L_Returns = L_Returns.rename(columns = {"Close": "Loblaws"})
Otex_Returns = Otex_Returns.rename(columns = {"Close": "Otex"})
```

```python
# Combine all stocks in single data frame and remove Symbol columns
Custom_Portfolio = pd.concat([Shop_Returns, L_Returns, Otex_Returns],axis='columns', join='inner')
Custom_Portfolio.drop(["Symbol"], axis="columns", inplace=True)
Custom_Portfolio.head()
```

```python
Daily_Returns_Custom_Portfolio = Custom_Portfolio.pct_change()
Daily_Returns_Custom_Portfolio
```

```python
Daily_Returns_Custom_Portfolio.isnull().sum()
```

```python
Daily_Returns_Custom_Portfolio.dropna(inplace=True)
```

```python
# Display Data
Daily_Returns_Custom_Portfolio.plot(figsize=(20,10), title = "Daily Returns")
```

```python
# Set weights
weights = [1/3, 1/3, 1/3]
Custom_Portfolio_Weighted_Returns = Daily_Returns_Custom_Portfolio.dot(weights)
Custom_Portfolio_Weighted_Returns
Custom_Portfolio_Weighted_Returns=Custom_Portfolio_Weighted_Returns.rename("Custom")
```

```python
# Join Custom Portfolio to Original Portfolio
All_Portfolios = pd.concat([Combine_Portfolio_Daily_Returns, Custom_Portfolio_Weighted_Returns], axis='columns', join='inner') 
All_Portfolios
```

```python
# Check for nulls
All_Portfolios.isnull().sum()
```

```python
All_Portfolios.plot(figsize=(20,10), title = "Daily Returns")
```

```python
# Calculate STD
All_Portfolios.std()
```

```python
# Calculate annualized STD
Annual_STD_ALL_Portfolios = All_Portfolios.std()*math.sqrt(252)
Annual_STD_ALL_Portfolios
```

```python
All_Portfolios.rolling(window=21).std()
```

```python
All_Portfolios.rolling(window=21).std().plot(figsize=(20,10), title = '21 Day Window')
```

```python
All_Portfolios_Correlation = All_Portfolios.corr()
All_Portfolios_Correlation
```

```python
sns.heatmap(All_Portfolios_Correlation, annot=True)
```

```python
# Calculate the variance of S&P TSX 60 (60-day window)
Rolling_SP_60Day_Variance = All_Portfolios['S&P TSX 60 Index'].rolling(window=60).var()
```

```python
#Calculate 60 covariance for Custom Portfolio
Custom_Covariance = All_Portfolios["Custom"].rolling(window=60).cov(All_Portfolios['S&P TSX 60 Index'])
```

```python
Beta_Custom = Custom_Covariance / Rolling_SP_60Day_Variance
Beta_Custom.plot(figsize=(20,10), title='Custom Beta')
```

```python
# Calculate Sharpe Ratio
All_Sharpe_Ratio = All_Portfolios.mean() / All_Portfolios.std()
All_Sharpe_Ratio
```

```python
# Calcualte Annual Shapre Ratios
Annual_All_Sharpe_Ratio = All_Sharpe_Ratio * math.sqrt(252)
Annual_All_Sharpe_Ratio
```

```python
# Visualize Annualized Sharpe Variables as a bar graph

Annual_All_Sharpe_Ratio.plot(kind='bar', title="Annualized Sharpe Ratios")
```

```python
# Which Portfolios Outperformed the S&P based on the Sharpe Ratio Values
Outperform_SP = Annual_All_Sharpe_Ratio > Annual_All_Sharpe_Ratio ["S&P TSX 60 Index"]
Outperform_SP
```
